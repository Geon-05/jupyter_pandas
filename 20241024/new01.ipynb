```python
df = sns.load_dataset('titanic')
male_mean = df[df['sex']=='male']['age'].mean()
female_mean = df[df['sex']=='female']['age'].mean()
df.loc[df['sex'] == 'male', 'age'] = df.loc[df['sex'] == 'male', 'age'].fillna(float(male_mean))
df.loc[df['sex'] == 'female', 'age'] = df.loc[df['sex'] == 'female', 'age'].fillna(float(female_mean))
```

```python
assert (df['age'].isnull().sum() == 0)
assert df['age'].mean().round(5) == 29.73603
```
