```python
from IPython.display import Image
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sample = df.copy()
sample.head()
```

```python
sample['class'].value_counts()
```

```python
def change_class(data_class):
    if data_class == 'First':
        return '일등석'
    elif data_class == 'Second':
        return '이등석'
    elif data_class == 'Third':
        return '삼등석'
```

```python
sample['class'] = sample['class'].apply(change_class)
sample['class'].value_counts()
```
