```python
import pandas as pd
import os
import pickle
```

```python
path = './data/네이버지식인_취업'
dir_list = os.listdir(path)
dir_list
```

```python
res= []
for file_name in dir_list:
    base_csv = pd.read_csv(f'./data/네이버지식인_취업/{file_name}')
    tmp_df = pd.DataFrame(base_csv)
    res.append(tmp_df)
df_all = pd.concat(res, ignore_index=True)
df_all
```

```python
with open('./data/네이버지식인_취업/df_data.pickle', 'wb') as f:
    pickle.dump(df_all, f, pickle.HIGHEST_PROTOCOL)
```
