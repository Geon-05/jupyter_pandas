```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터 시각화툴
# - power bi
# - 테블로
# - seaborn

# seaborn에서 제공하는 기본 데이터셋들중 'tips'를 사용해보자!
df = sns.load_dataset('tips')
```

```python
# 신뢰구간 - 하단그래프에서 검은색 선 구간
# 전체데이터(모수)의 평균범위를 표현하는 구간
# 데이터추정치는 표본으로 이루어지기때문에!
# errorbar=95 -> 기본값 5% 오차범위를 나타냄
# hue='sex' -> 기준데이터를 세부표현함 / 일반적으로는 문자열로
# palette='coolwarm' -> 컬러팔레트 / 도큐먼트 참고!

# 바플롯
# sns.barplot(x='x축데이터', y='y축데이터', data='데이터프레임', **속성)
# 바이올린플롯
# sns.violinplot(x='x축데이터', y='y축데이터', data='데이터프레임', **속성)
# 박스플롯
# sns.boxplot(x='x축데이터', y='y축데이터', data='데이터프레임', **속성)
# 산점도 - x,y축데이터 숫자일것!
# sns.relplot(x='x축데이터', y='y축데이터', data='데이터프레임', **속성)
# height=5 -> x축데이터사이즈, aspect=1 -> y축데이터사이즈 
# 카운트플롯
# sns.countplot(x='x축데이터', data='데이터프레임', **속성)
# 히트맵
# sns.heatmap('데이터프레임')

sns.barplot(x='day', y='total_bill', data=df)
```

```python
sns.barplot(x='day', y='total_bill', data=df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, hue='smoker')
```
