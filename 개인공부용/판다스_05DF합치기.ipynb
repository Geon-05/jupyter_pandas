```python
import pandas as pd
import os

print(os.getcwd())

아이돌1 = pd.read_csv('./data/korean-idol.csv')
아이돌2 = pd.read_csv('./data/korean-idol-2.csv')
```

```python
# 데이터 프레임 붙이기 - concat
한국아이돌 = pd.concat([아이돌1, 아이돌2])
display(한국아이돌)
한국아이돌 = pd.concat([아이돌1, 아이돌2], ignore_index=True)
display(한국아이돌)

'''
🔍 pd.concat(데이터프레임리스트) : 데이터프레임을 물리적으로 이어 붙여준다.
🤔 기본값으로 axis=0 이므로 위아래로 이어붙여주게 된다.
이때, 존재하지 않는 값은 NaN이된다.
기본값으로 ignore_index=False 이므로 기존의 인덱스를 그대로 가져온다.
따라서, 인덱스를 새로 정의하려면 ignore_index=True 하자!
'''
```

```python
한국아이돌 = pd.concat([아이돌1, 아이돌2], axis=1)
display(한국아이돌)
'''
🤔 axis=1 로 바꾸면 옆으로 이어붙여주게 된다.
concat함수는 기본값으로 join='outer'(합집합)이다.
교집합을 원한다면 join='inner' 속성을 주자!
'''
```

```python
# 데이터프레임 병합 - merge
한국아이돌 = pd.merge(아이돌1, 아이돌2)
display(한국아이돌)
한국아이돌 = pd.merge(아이돌1, 아이돌2, how='outer', on='이름')
display(한국아이돌)
'''
🔍 pd.concat(df_left, df_right, how='inner', on=None)
🤔 데이터프레임을 각 데이터에 존재하는 고유값(key)을 기준으로 병합한다.
how='left'  : 왼쪽데이터프레임 기준
how='right' : 오른쪽데이터프레임 기준
how='inner' : 교집합
how='outer' : 합집합
'''
```
