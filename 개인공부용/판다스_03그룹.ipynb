```python
import pandas as pd
import os

print(os.getcwd())

데이터프레임 = pd.read_csv('./data/house_price_clean.csv')
```

```python
# 그룹화
display(데이터프레임.groupby('지역'))
지역그룹 = 데이터프레임.groupby('지역')
연도그룹 = 데이터프레임.groupby('연도')
'''
🔍 데이터프레임.groupby('열 이름')
🤔 '데이터프레임'데이터프레임을 '열 이름'컬럼으로 그룹화 한 상태로 주소가 반환된다.
<pandas.core.groupby.generic.DataFrameGroupBy object at 0x000001997754F980>
해당 '열 이름'컬럼은 그룹화된 데이터프레임의 인덱스가 된다.
따라서, 이후에 집계함수를 이용하여 처리를 진행해야한다.

아래에서는 한글로 그룹화한 지역그룹과 정수로 그룹화한 연도그룹으로 연습해볼것이다!

count() : 데이터의 개수를 세는 함수
sum() : 데이터의 합을 구하는 함수
mean() : 데이터의 평균을 구하는 함수
median() : 데이터의 중앙값을 구하는 함수
min() : 데이터의 최소값을 구하는 함수
max() : 데이터의 최대값을 구하는 함수
std() : 데이터의 표준편차를 구하는 함수
var() : 데이터의 분산을 구하는 함수
'''
```

```python
# count : 데이터의 개수를 세는 함수
display(지역그룹.count())
display(연도그룹.count())
display(지역그룹.count().loc['강원'])
display(연도그룹.count().loc[2015])
display(지역그룹['규모'].count())
display(연도그룹['규모'].count())
'''
🔍 그룹화된 각각의 인덱스의 데이터의 개수를 세어준다.
🤔 loc를 활용하면 특정 열의 자료를 얻을 수 있다.
'''
```

```python
# sum : 데이터의 합을 구하는 함수
display(지역그룹.sum())
display(연도그룹.sum())
display(지역그룹.sum().loc['강원'])
display(연도그룹.sum().loc[2015])
display(지역그룹['분양가'].sum())
display(연도그룹['분양가'].sum())
'''
🔍 그룹화된 각각의 인덱스의 데이터의 합을 산출한다.
🤔 문자열은 오류가 나지 않으나 결과를 보아하니 피해야겠다..
'''
```

```python
# mean : 데이터의 합을 구하는 함수
# display(지역그룹.mean()) -> TypeError
# display(연도그룹.mean()) -> TypeError
# display(지역그룹.mean().loc['강원']) -> TypeError
# display(연도그룹.mean().loc[2015]) -> TypeError
display(지역그룹['분양가'].mean())
display(연도그룹['분양가'].mean())
'''
🔍 그룹화된 각각의 인덱스의 데이터의 평균을 산출한다.
🤔 문자열은 타입오류를 발생시킨다.
'''
```
