```python
test_data = pd.read_csv('./혼밥족_Data/연도별 세대원수/2014_세대원수별 세대수.csv', encoding='cp949')
test_data['행정구역'] = test_data['행정구역'].map(lambda x : x.split('(')[1].replace(')',''))
test_data = test_data.drop(0)
test_data = test_data[~test_data['행정구역'].str.contains('00000')].reset_index(drop=True).T
test_data = test_data.reset_index()
test_data = test_data[~test_data['index'].str.contains('전체세대')].reset_index(drop=True)
for i in range(1,13):
    tmp = test_data[test_data['index'].str.contains(f'{i:02}월')]
    tmp.columns = test_data.loc[0]
    기준일 = tmp.iloc[0,0][:8]
    tmp['행정구역'] = tmp['행정구역'].map(lambda x : x[9:])
    tmp = tmp.T
    tmp = tmp.reset_index()
    tmp['기준일'] = 기준일
    display(tmp)
# test_data = test_data[~test_data['index'].str.contains('전체세대')].reset_index(drop=True).T
# test_data
    
```
