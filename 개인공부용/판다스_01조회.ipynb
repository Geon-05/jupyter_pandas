```python
# 열 선택하기
display(변수명['지역'])
display(변수명[['지역','규모']])
'''
🔍 변수명['열 이름']
🤔 특정한 열을 선택할 때
🔍 변수명[['열 이름1', '열 이름2', ...]]
🤔 여러개의 열을 선택할 때 단, 리스트 형태로 열 이름을 전달한다!!

깔끔한 코딩을 하려면 열 이름의 리스트를 변수를 사용해보자!
col = ['지역', '규모']
display(변수명[col])
'''
```

```python
# 행 선택하기 - loc
display(변수명.loc[0])
display(변수명.loc[[0,1]])
display(변수명.loc[:2])
'''
🔍 변수명.loc[행 이름 또는 인덱스]
🤔 특정한 행을 선택할 때
🔍 변수명[['행 이름 또는 인덱스1', '행 이름 또는 인덱스2', ...]]
🤔 여러개의 열을 선택할 때 단, 리스트 형태로 열 이름을 전달한다!!
🔍 변수명[슬라이싱]
🤔 슬라이싱도 가능!
'''

# 행 선택하기 - iloc / 인덱스를 넣는 자리에는 정수만 입력가능하다!
display(변수명.iloc[0])
display(변수명.iloc[0,2])
display(변수명.iloc[:2,:2])
'''
🔍 변수명.iloc[행 인덱스]
🤔 특정한 행을 선택할 때
ex) 변수명.loc[0]과 같은 효과 / 변수명.iloc[0,:]도 같음
🔍 변수명.iloc[행 인덱스,열 인덱스]
🤔 특정 데이터를 선택할 때
ex) 변수명.iloc[0,2] -> 0번째 행의 2번째 열의 데이터값 추출
🔍 변수명.iloc[슬라이싱,슬라이싱]
🤔 특정 구간을 선택할 때
'''

'''
🤔 loc는 칼럼명을 직접 적거나 특정 조건식을 써줌으로써
사람이 읽기 좋은 방법으로 데이터에 접근.
🤔 iloc는 컴퓨터가 읽기 좋은 방법으로 데이터가 있는
위치를 정수로(순서) 접근.
'''
```

```python
# 데이터 필터링
display(변수명[변수명['연도']==2015])
'''
🔍 변수명[조건]
🤔 특정 조건에 해당하는 데이터를 선택하는 방법
ex) 변수명[변수명['연도']==2015] -> 변수명 데이터 프레임에서 '연도'컬럼에서 2015만 필터링
'''
display(변수명[변수명['연도'].isin([2015])])
'''
🔍 변수명[변수명[열 이름].isin(찾고자하는리스트)]
🤔 특정 조건에 해당하는 데이터를 선택하는 방법 / 리스트로 넣어야한다!
'''
```

```python
# 데이터 정렬하기
display(변수명.sort_values('분양가'))
display(변수명.sort_values('분양가', ascending=False))
'''
🔍 변수명.sort_values('열 이름')
🤔 오름차순 : ascending=True (기본값) / 내림차순 : ascending=False
'''
```

```python
# 인덱스 초기화 및 설정
display(변수명.reset_index())
변수명 = 변수명.reset_index()
'''
🔍 변수명.reset_index()
🤔 인덱스를 초기화 하고 기존 인덱스는 새로운 열로 추가된다.
'''

display(변수명.set_index('index'))
'''
🔍 변수명.set_index('컬럼명')
🤔 특정 컬럼을 인덱스로 설정한다.
'''
```
